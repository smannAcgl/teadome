```python
from kingdon import Algebra

alg = Algebra(3, 0, 1)
locals().update(alg.blades)
```

```python
# Construct points and lines from coefficients
point = lambda x, y, z: (e0 + x*e1 + y*e2 + z*e3).dual()
plane = lambda a, b, c, d: a*e1 + b*e2 + c*e3 + d*e0
plane_thru_point = lambda n, pt: n + (pt & n)*e0
normal_from_point = lambda pt, e0:   dual() & pt
```

```python
# projection function
project = lambda A, p: (A | p)/p   
```

```python
A = point(0,1,0)
B = point(2,0,0)
# p = plane_thru_point(1*e1,B)  # only here makes it static
```

```python
def graph_func():
    p = plane_thru_point (1*e1, B)
    Ap = project(A,p)
    pA = project(p,A)
    
    return [                                        
        0x00AA88, p, "p", pA, "pA", 
        0xff0000, A, "A", Ap, "Ap",
        0x0000ff, B, "B",
    ]
    
```

```python
alg.graph(
    graph_func,
    lineWidth=3, 
    grid=1, 
    labels=1, 
    h=0.6, 
    p=-0.15, 
    pointRadius=2,
    fontSize=1,
    scale=1,
)
```
